## Part 4: Trend Analysis and Seasonality

### Trend Analysis (Moving averages)

<div style="direction:rtl; text-align:right; font-size: 16px; color:blue">
     برای تحلیل روند از تکنیک میانگین متحرک به صورت ۳۰روزه (ماهانه) و ۱۸۰روزه (فصلی) استفاده کردیم.
</div>

```python
import matplotlib.pyplot as plt

def plot_moving_average(data, data_name, window_size=30,window_size2=60):
    name = data_name
    # Compute the moving average
    data['MA'] = data['Close'].rolling(window=window_size).mean()
    data['MA2'] = data['Close'].rolling(window=window_size2).mean()
    # Plot the original 'Close' prices and the moving average
    plt.figure(figsize=(12, 6))
    plt.plot(data.index, data['Close'], label='Original')
    plt.plot(data.index, data['MA'], label=f'{window_size}-day Moving Average')
    plt.plot(data.index, data['MA2'], label=f'{window_size2}-day Moving Average')
    plt.title(f'{name} Close Prices with Moving Average')
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.show()
```

```python
plot_moving_average(apple_data,"Apple Inc. Stock",30,180)
```

<div style="direction:rtl; text-align:right; font-size: 15px; color:blue">
     میانگین متحرک سهام شرکت اپل در طول زمان به طور عمده حرکتی به سمت بالا داشته به جز بازه‌ی زمانی کوتاهی در سال ۲۰۱۹ و ۲۰۲۱ و روند نامنظمی در سال ۲۰۲۲ داشته و در نهایت در همین سال سهام به بیشترین مقدار خود می‌رسد.
</div>

```python
plot_moving_average(microsoft_data,"Microsoft Corporation stock",30,180)
```

<div style="direction:rtl; text-align:right; font-size: 15px; color:blue">
     میانگین متحرک سهام شرکت مایکروسافت نیز در طول زمان به طور عمده حرکتی به سمت بالا داشته به جز بازه‌ی زمانی کوتاهی در سال ۲۰۲۱ و در نهایت در سال ۲۰۲۲ به بیشترین مقدار خود می‌رسد و پس از آن روندی نزولی را در پیش می

### Seasonality

<div style="direction:rtl; text-align:right; font-size: 15px; color:black">
     در ادامه برای هرکدام از شرکت‌ها سه نمودار نمایش داده می‌شود. که نمودار اول نمایش فصلی سهام شرکت و نشان دهنده‌ی یک الگوی تکراری است. نمودار دوم روند ماه‌های سال در طول زمان است و نمودار سوم روند هرسال بر اساس ماه‌های مختلف را نشان می‌دهد. 
</div>

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(apple_data['Close'], model='additive', period=180)
plt.figure(figsize=(15, 5))
plt.title('Seasonality Analysis of Apple Inc. Stock Close Prices')
result.seasonal.plot()
plt.show()
```

<div style="direction:rtl; text-align:right; font-size: 15px; color:blue">
     در نمودار بالا الگوهای تکراری را می‌توان در طول زمان مشاهده کرد که نشان دهنده‌ی فصلی بودن آن است. اما چون این الگو برای زمانی بین یک تا دو سال می‌باشد، روند فصلی بودن و الگوی تکراری در دو نمودار بعدی دیده نمی‌شود.
</div>

```python
import matplotlib.pyplot as plt

# Extract the month and day from the index
apple_data['Month'] = apple_data.index.month
apple_data['Year'] = apple_data.index.year

# Calculate the average closing price by month and day
average_price = apple_data.groupby(['Year', 'Month'])['Close'].mean()

# Reshape the data for plotting
average_price = average_price.unstack()

# Plot the seasonality pattern
plt.figure(figsize=(50, 50))
average_price.plot()
plt.title('Seasonality Analysis of Apple Inc. Stock Close Prices')
plt.xlabel('Year')
plt.ylabel('Average Close Price')
plt.legend(title='Month')
plt.show()
```

<div style="direction:rtl; text-align:right; font-size: 15px; color:blue">
     به طور کلی روند صعودی برای هر ماه به طور جداگانه مشاهده می‌شود.اما می‌توان مشاهده کرد که در ماه‌های مختلف سال روند سهام شرکت اپل افت و خیزهایی داشته است.
